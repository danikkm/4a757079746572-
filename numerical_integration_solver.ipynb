```python
# Midpoint
def Midpoint(f, a, b, n):
    h = float(b - a) / n
    
    result = 0
    for i in range(n):
        result += f((a + h / 2.0) + i * h)
    result *= h
    return result


def Integral(f):
    return quad(f, lower_limit, upper_limit)


def AbsoluteError(integral, midpointIntegral):
    return np.absolute(integral[0] - midpointIntegral)


def Derivative(x, y):
    return y.diff(x, 2)


def TheoreticalError(a, b, x, y, n):
    derivative2nd = Derivative(x, y)
    function = lambdify(x, derivative2nd, "numpy")
    M = int(function(np.ones(1)))
    h = float(b - a) / n

    return (h ** 2) * (b - a) / 24 * M


def RungeError(midpointIntegral, midpointIntegral2N):
    return np.absolute((midpointIntegral2N - midpointIntegral)) / ((2 ** 2) - 1)


# Main
"""
x - symbol for sympy derivative only
n - number of iterations
lower_limit - lower limit, otherwise a 
upper_limit - upper limit, otherwise b
y - function for sympy derivative only
"""
print(" Midpoint integral \n")
x = Symbol("x")

n = 100
lower_limit = 0
upper_limit = 1
y = exp(x**2)


# Vidurio tasko integralas

"""
1st argument: f - function
2nd argument: a - beginning of the interval
3rd argument: b - end of the intreval 
4th argument: n - number of iterations

"""
midpointIntegral = Midpoint(lambda x: exp(x**2), lower_limit, upper_limit, n)
midpointIntegral2N = Midpoint(lambda x: exp(x**2), lower_limit, upper_limit, 2 * n)

print(" S =", "%.15f" % midpointIntegral)

# Integralas

"""
1st argument: f - function
"""
integral = Integral(lambda x: exp(x**2))
print(" Integral - ", integral[0])

# Absoliutine realioji paklaida

"""
1st argument: integral - integral 
2nd argument: midpointIntegral - middle point integral
"""

print(" E = ", "%.15f" % (AbsoluteError(integral, midpointIntegral)))

# Isvestine

"""
1st argument: x - sympy symbol for derivative
2nd argument: y - sympy function for derivative
"""
derivative = Derivative(x, y)

# Teorinis paklaidos ivertis

"""
1st argument: a - beginning of the interval
2nd argument: b - end of the intreval 
3rd argument: x - sympy symbol for derivative
4th argument: y - sympy function for derivative
5th argument: n - number of iterations
"""

print("|R|(paklaida) =", "%.15f" % (TheoreticalError(lower_limit, upper_limit, x, y, n)))

# Runges taisykle
"""
1st argument: midpointIntegral
2nd argument: 2 * midpointIntegral 
"""
print("|R2N|", "%.15f" % (RungeError(midpointIntegral, midpointIntegral2N)))
```

```python
# Simpson
def Simpson(f, a, b, n):

    h = float(float(b) - float(a)) / n
    k = 0.0
    x = a + h
    for i in range(1, n // 2 + 1):
        k += 4 * f(x)
        x += 2 * h

    x = a + 2 * h
    for i in range(1, n // 2):
        k += 2 * f(x)
        x += 2 * h

    return (h / 3) * (f(a) + f(b) + k)


def Integral(f):
    return quad(f, lower_limit, upper_limit)


def AbsoluteError(integral, simpsonIntegral):
    return np.absolute(integral[0] - simpsonIntegral)


def Derivative(x, y):
    return y.diff(x, 4)


def TheoreticalError(a, b, x, y, n):
    derivative4nd = Derivative(x, y)
    # print(derivative4nd, 'derivative')
    function = lambdify(x, derivative4nd, "numpy")
    M = int(function(np.ones(1)))
    # print(M, "MAXI")
    h = float(b - a) / n

    return (h ** 4) * (b - a) / 2880 * M


def RungeError(simpsonIntegral, simpsonIntegral2N):
    return np.absolute((simpsonIntegral2N - simpsonIntegral)) / ((2 ** 4) - 1)


# Main

"""
x - symbol for sympy derivative only
n - number of iterations
lower_limit - lower limit, otherwise a 
upper_limit - upper limit, otherwise b
y - function for sympy derivative only
"""

print(" Simpson integral \n")
x = Symbol("x")

n = 4 # mb + 1 
lower_limit = 0
upper_limit = 4
y = 3*x**4+4*x-5


# Simpsono  integralas

"""
1st argument: f - function
2nd argument: a - beginning of the interval
3rd argument: b - end of the intreval 
4th argument: n - number of iterations

"""
simpsonIntegral = Simpson(lambda x: 3*x**4+4*x-5, lower_limit, upper_limit, n)
simpsonIntegral2N = Simpson( lambda x: 3*x**4+4*x-5, lower_limit, upper_limit, 2 * n)

print(" S =", "%.15f" % (simpsonIntegral))

# Integralas

"""
1st argument: f - function
"""
integral = Integral(lambda x:  3*x**4+4*x-5)
print(" Integralas - ", integral[0])

# Absoliutine realioji paklaida

"""
1st argument: integral - integral 
2nd argument: midpointIntegral - simpson integral
"""

print(" E = ", "%.15f" % (AbsoluteError(integral, simpsonIntegral)))

# Isvestine

"""
1st argument: x - sympy symbol for derivative
2nd argument: y - sympy function for derivative
"""
derivative = Derivative(x, y)

# Teorinis paklaidos ivertis

"""
1st argument: a - beginning of the interval
2nd argument: b - end of the intreval 
3rd argument: x - sympy symbol for derivative
4th argument: y - sympy function for derivative
5th argument: n - number of iterations
"""

print("|R| =", "%.15f" % (TheoreticalError(lower_limit, upper_limit, x, y, n)))

# Runges taisykle
"""
1st argument: simpsonIntegral
2nd argument: 2 * simpsonIntegral 
"""
print("|R2N|", "%.15f" % (RungeError(simpsonIntegral, simpsonIntegral2N)))
```
