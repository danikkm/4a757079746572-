```python
#MINIMUM
from scipy.constants import golden

 
def goldensearch(f,a,b,tol):
    xl = b - (b-a)/golden
    xu = a + (b-a)/golden
    
    
    eps = abs(b-a)

    iteration = 2
    while eps > tol:
        # print a,b,eps
        if f(xl)<f(xu):
            b = xu
            xu = xl
            xl = b - (b-a)/golden
           
        else:
            a = xl
            xl = xu
            xu = a + (b-a)/golden
        print(iteration, 'Iteration,', 'xl = ', xl,'\n')
        print(iteration, 'Iteration,', 'xu =', xu, '\n')
        print(iteration, 'Iteration,', "Minimum = ", min(xu,xl))
        print("_________________________________________", '\n')
            
        eps = abs(b-a)
        iteration += 1
        if (iteration > 1000):
            print('Enough?')
            break;

    return (xu+xl)/2

f = lambda x: x**3-4*x**2
print(goldensearch(f,0,3,1e-5), '\n')

```

```python
def newton(f,Df,x0,epsilon,max_iter):
    
    if max_iter == None:
        max_iter = 1000
        
    xn = x0
    for n in range(0,max_iter):
        fxn = f(xn)
        if abs(fxn) < epsilon:
            print('Found solution after',n,'iterations.')
            return xn
        Dfxn = Df(xn)
        if Dfxn == 0:
            print('Zero derivative. No solution found.')
            return None
        xn = xn - fxn/Dfxn
    print('Exceeded maximum iterations. No solution found.')
    return None

f = lambda x: ((x**3-4*x**2))
Df = lambda x: (3*x**2 - 8*x)
approx = newton(f,Df,-2,1e-10,None)
print(approx)
```

```python
def dx(f, x):
    return abs(0-f(x))

def newtons_method(f, df, x0, e):
    delta = dx(f, x0)
    while delta > e:
        x0 = x0 - f(x0)/df(x0)
        delta = dx(f, x0)
    print ('Root is at: ', x0)
    print ('f(x) at root is: ', f(x0))
    
    
def f(x):
    return (x**3-3*x**2+1)/(x**3)
 
def df(x):
    return (3*x**2 - 6*x)/x**3 - 3*(x**3 - 3*x**2 + 1)/x**4

newtons_method(f, df, 0.9, 1e-4)
```
