```python
#PUSIAUKIRTOS METODAS
"""Bisection"""
import numpy as np
import math

def samesign(a, b):
    return a * b > 0

def bisect(func, low, high, precision = None):

    "Find root of continuous function where f(low) and f(high) have opposite signs"
    assert not samesign(func(low), func(high))
    count = 0
    
    for i in range(1000):
        midpoint = (low + high) / 2.0
        length = (high - low) / 2.0
        if samesign(func(low), func(midpoint)):
            low = midpoint
            #print("Low is here:", low, samesign(func(low), func(midpoint)))

        else:
            high = midpoint
            #print("High is here:", high, samesign(func(low), func(midpoint)))

        if precision is not None:
            if length < precision:
                break
        
        count = count + 1
    
    print(count)
    # print(midpoint)

    return midpoint


def f(x):
    return x**3 - 2 - x

func = bisect(f, 0, 2, 10e-2)

x = Symbol("x")
y = sin(x) - x

print("Bisection:", func)
print("Solution = ", solveset(y, x, Interval(math.pi/2, math.pi/4)))

```

```python
#PUSIAUKIRTOS METODAS
"""Bisection"""
def bisection(f,a,b,N):
    '''Approximate solution of f(x)=0 on interval [a,b] by the bisection method.

    Parameters
    ----------
    f : function
        The function for which we are trying to approximate a solution f(x)=0.
    a,b : numbers
        The interval in which to search for a solution. The function returns
        None if f(a)*f(b) >= 0 since a solution is not guaranteed.
    N : (positive) integer
        The number of iterations to implement.

    Returns
    -------
    x_N : number
        The midpoint of the Nth interval computed by the bisection method. The
        initial interval [a_0,b_0] is given by [a,b]. If f(m_n) == 0 for some
        midpoint m_n = (a_n + b_n)/2, then the function returns this solution.
        If all signs of values f(a_n), f(b_n) and f(m_n) are the same at any
        iteration, the bisection method fails and return None.

    Examples
    --------
    >>> f = lambda x: x**2 - x - 1
    >>> bisection(f,1,2,25)
    1.618033990263939
    >>> f = lambda x: (2*x - 1)*(x - 3)
    >>> bisection(f,0,1,10)
    0.5
    '''
    if f(a)*f(b) >= 0:
        print("Bisection method fails.")
        return None
    a_n = a
    b_n = b
    for n in range(1,N+1):
        m_n = (a_n + b_n)/2
        f_m_n = f(m_n)
        if f(a_n)*f_m_n < 0:
            a_n = a_n
            b_n = m_n
        elif f(b_n)*f_m_n < 0:
            a_n = m_n
            b_n = b_n
        elif f_m_n == 0:
            print("Found exact solution.")
            return m_n
        else:
            print("Bisection method fails.")
            return None
        print("N =",n, "an =",a_n," bn=",b_n, "cn =", (a_n + b_n)/2, '(bn - an) / 2 =', (b_n- a_n) /2)
        
    print()
    return (a_n + b_n)/2

x = Symbol("x")
y = exp(x) - 3*x

f = lambda x: math.exp(x) - 3*x
print("Bisection:",bisection(f,0,1,20))

print("Solution = ", solveset(y, x, Interval(0, 1)))

```

```python
#Niutono metodas
"""Newton"""
from sympy import *

def Derivative(x, y):
    return y.diff(x, 1)

def newton(f,Df,x0,epsilon,max_iter):
    xn = x0
    for n in range(0,max_iter):
        fxn = f(xn)
        if abs(fxn) < epsilon:
            print('Found solution after',n,'iterations.')
            return xn
        Dfxn = Df(xn)
        if Dfxn == 0:
            print('Zero derivative. No solution found.')
            return None
        xn = xn - fxn/Dfxn
    print('Exceeded maximum iterations. No solution found.')
    return None



###
#For sympy derivative
x = Symbol("x")
y = x**3 - 2*x**2 -1 - 3*x
###

f = lambda x:  x**3 - 2*x**2 -1 - 3*x
Df = lambda x: 3*x**2 - 4*x - 3

print(Derivative(x, y))

""" 
Parameters
    ----------
    f : function
        Function for which we are searching for a solution f(x)=0.
    Df : function
        Derivative of f(x).
    x0 : number
        Initial guess for a solution f(x)=0.
    epsilon : number
        Stopping criteria is abs(f(x)) < epsilon.
    max_iter : integer
        Maximum number of iterations of Newton's method.
"""
print(newton(f,Df,3,1e-4,10))
```

```python
#Kirstiniu metodas
"""Secant"""
def secant(f, x0, x1, eps):
    f_x0 = f(x0)
    f_x1 = f(x1)
    iteration_counter = 0
    while abs(f_x1) > eps and iteration_counter < 100:
        try:
            denominator = float(f_x1 - f_x0)/(x1 - x0)
            x = x1 - float(f_x1)/denominator
        except ZeroDivisionError:
            print ("Error! - denominator zero for x = ", x)
            sys.exit(1)     # Abort with error
        x0 = x1
        x1 = x
        f_x0 = f_x1
        f_x1 = f(x1)
        iteration_counter += 1
    # Here, either a solution is found, or too many iterations
    if abs(f_x1) > eps:
        iteration_counter = -1
    return x, iteration_counter

f = lambda x: exp(-2*x) - x**2

solution, no_iterations = secant(f, 0, 3, eps=1.0e-6)

if no_iterations > 0:    # Solution found
    print ("Number of function calls: %d" % (2 + no_iterations))
    print ("A solution is: %f" % (solution))
else:
    print ("Solution not found!")
```

```python
import math

def fixed_point(f,init,N=100):
    """
    Implementation of the fixed-point method
    Parameters:
        f    - Function to use
        init - Initialization for the algoirthm
        N    - Number of iterations
    Return value:
        p - An array containing the iterations of the method
    """

    p = [init]
    for i in range(0,N):
        p.append(f(p[i]))

    return p

def f(x):
    return 2**-x

# Start iteration
p = fixed_point(f,1,10)
for i in range(0,len(p)):
    if p[i] > 0:
        print(f"{i}  {p[i]} {abs(p[i + 1] - p[i])}")
    else:
        print(f"{i} {p[i]} {abs(p[i + 1] - p[i])}")
        
```
